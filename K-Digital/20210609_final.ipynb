### 1. 전국별로 성별 가져오기

```python
# 시군구별이 전국인 것만 df_all 에 넣는다.
df_all = df[(df['시군구별']=='전국')]
df_all.head()
```

```python
# 이 중에서 성별이 전체인 것을 가져온다.
df_all = df[(df['시군구별']=='전국') & (df['성별']=='전체')]
df_all.head()
```

```python
# head 와 tail 데이터로 기간 알아본다
# 1997년 부터 2019년 데이터까지 있구나~~~~ 
df_all.tail()
```

```python
# 이제 확인해보자
# 시군구별 전체만 가져왔는지, 성별 전체만 가져왔는지 확인하는 작업
# 아래 데이터는 비교 대상이 다르기 떄문에 (시군구별, 성별) isin 사용 불가!!!
# 같았으면 isin 사용 가능
df_all = df[(df['시군구별']=='전국') & (df['성별']=='전체')]
print(df_all['시군구별'].unique())
print(df_all['성별'].unique())
```

## 각 지역별의 전체만 가져와서 살펴보자!!!

```python
df_local_all = df_local[df_local['성별'] == '전체']
df_local_all.head()
```

```python
# 전체만 가져와서 그래프를 그려보자!!
plt.figure(figsize = (15, 4))
sns.pointplot(data = df_local_all, x = '연도', y = '출생아수')
```

```python
# 전국, 시군구별로 나눠서 보자!!!
# 전체만 가져와서 그래프를 그려보자!!
plt.figure(figsize = (15, 4))
sns.pointplot(data = df_local_all, x = '연도', y = '출생아수', hue='시군구별')

# bbox_to_anchor = (x, y, width, height)
# ncol = 1
plt.legend(loc='center right', bbox_to_anchor=(1.17, 0.5), ncol=1)
# 범례 오른쪽으로 옮기자 (loc = 'center right'), 그래프랑 범례랑 겹친다...(bbox_to_anchor=(1.17, 0.5), ncol=1)
```
