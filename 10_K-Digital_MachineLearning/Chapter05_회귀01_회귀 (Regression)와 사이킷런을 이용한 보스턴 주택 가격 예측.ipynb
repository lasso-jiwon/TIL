# 회귀(Regression)
분류와 더불어 지도학습의 가장 큰 축 중의 하나   
사실 분류보다 회귀가 현업 업무 분석가들이나 데이터 분석가들이 많이 활용  
왜냐하면 대부분 기업에서 사용하는 예측이 수치적 예측이 많기 때문에  
회귀가 지도학습에서 가장 많이 활용되는 분야  
현대 통계학을 이루는 큰 축이고, 거의 모든 분야에서 회귀가 많이 사용되고 있음  
우리는 머신러닝 관점에서 회귀 사용  

회귀(Regression)란   
여러 개의 독립변수와 한 개의 종속변수 간의 상관관계를 모델링하는 기법  
Y = W1X1 + W2X2 + W3X3 + ... + WnXn  
- 독립 변수 (X) : 서로 간 상관관계가 적고 독립적인 변수  
    - 피처 : 아파트 크기, 방 개수, 주변 지하철역 수, 주변 학군 등
- 종속 변수 (Y) : 
    - 결정값 : 아파트 가격
- 회귀 계수 (W) : 독립 변수의 값에 영향을 미치는 것

머신러닝의 회귀 기법은 최적의 회귀 계수를 찾아 예측하는 방법임    
주어진 피처와 결정값 데이터로 학습을 통해 최적의 회귀계수를 찾아 회귀식을 만들어  
실제 테스트 데이터 셋에서 예측한다.  
피처값과 결정값 (종속값)은 이미 주어지고 학습을 통해서 최적의 회귀계수(W)를 찾는 것  
회귀에서 가장 중요한 것은 회귀 계수이다.  

회귀 유형  
- 선형/비선형 : 회귀 계수의 선형/비선형 여부에 따라 나눔
- 단일 회귀/다중(다항) 회귀 : 독립 변수의 개수에 따라 나눔

분류와 회귀 차이점 : 결과값의 차이  
분류 : 카테고리 값 반환 (이산값)
-0/1, 이미지1/이미지2, 카테고리값 0/1/2, negative/positive

## 선형 회귀  
가장 많이 사용되는 회귀  
실제값과 예측값의 차이 (오류의 제곱값)을 최소화 하는 직선형 회귀선을 최적화하는 방식  
실제값과 예측값의 차이인 오류를 최소로 줄일 수 있는 선형 함수를 찾아서  
이 선형 함수에 독립변수 (피처)를 입력해 종속변수(타깃값, 예측값)을 예측하는 것  
규제 방법에 따라 여러 유형으로 나뉨  
  
  
규제 : 일반적인 선형 회귀의 과적합 문제를 해결하기 위해 회귀 계수에 패널티 값을 적용하는 것  

## 대표적인 선형 회귀 모델
- **일반 선형 회귀** : 규제를 적용하지 않은 모델로 예측값과 실제값의 손실함수(RSS)를 최소화할 수 있도록 최적화 
- **릿지(Ridge)** : 선형 회귀에 L2규제를 추가한 회귀 모델
- **라쏘(Lasso)** : 선형 회귀에 L1 규제를 적용한 모델
- **엘라스틱넷(ElasticNet)** : L2, L1 규제를 함께 결합한 모델
- **로지스틱 회귀(Logistic Regression)** : 분류에서 사용되는 선형 모델

# 단순 선형 회귀를 통한 회귀 이해

독립변수(피처)가 하나뿐인 선형회귀  
예를들어 주택 가격이 단순히 주택 크기로만 결정된다는 단순 선형 회귀로 가정  
주택 가격은 주택 크기에 대해 선형(직선 형태)의 관계로 표현  
X축이 주택 크기 (평수)이고 Y축이 주택의 가격 축이라면 아래 그림과 같음

왼쪽 그래프는 전반적으로 주택 가격의 추세를 보면 주택 크기에 따라  
가격이 증가됨은 볼 수 있지만 반드시 그렇지많은 않음.  
직선(선형)표시로 기울기와 절편을 가진 1차 함수식으로 모델링 (예측 모델)  
  
오른쪽 그래프는 실제값과 모델 사이의 오류값을 표시함  
  
최적의 회귀 모델은 실제 값과 모델 사이의 오류값(잔차) 합이 최소가 되는 모델임  
오류값 합이 최소가 될 수 있는 최적의 회귀 계수를 찾는 것  
  
오류값은 +나 -가 될 수 있음. 그래서 전체 데이터의 오류 합을 구하기 위해 단순히 더했다가는 원치 않는 결과가 나오므로  
보통 오류의 합을 계산할 때는 절대값을 취하거나 오류값의 제곱을 구해서 더하는 방식(RSS)를 사용함  
학습을 통해 이 RSS를 최소로 하는 회귀 계수를 찾는 것이 머신러닝 기반의 회귀 핵심 사항  
RSS(Residual Sum of Square) : Error의 제곱 

머신러닝 회귀 알고리즘은 데이터를 계속 학습하면서 이 오류값을 감소시키고  
최종적으로는 더 이상 감소하지 않는 최소의 오류 값을 구하는 것

# 선형회귀 예제

```python
# 선형회귀 사이킷런 사용법
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 데이터 생성
np.random.seed()

# X축 값
X = 2 * np.random.rand(100, 1) # 100개

# Y축 값 : y = 6 + 4X (w0=6, w1=4)
# 퍼져 보이게 하기 위해 노이즈 값을 줌
y = 6 + 4*X + np.random.randn(100,1)

plt.scatter(X, y)
```

rand()   
- 0~1 사이에서 균일한 확률 분포로 실수 난수 생성
- 음수 값이 없음
  
randn()  
- 표준 정규 분포(평균이 0이고, 표준편차가 1인 분포)를 따르는 난수 발생
- 음수값이 나올 수 있음

```python
np.random.rand(10, 1)
```

```python
# 단순 선형 회귀 분석
# LinearRegression 알고리즘으로 학습
# X : 피처값
# Y : 타깃값

from sklearn.linear_model import LinearRegression
line_fitter = LinearRegression() # 객체 생성
line_fitter.fit(X, y)
```

```python
# X가 1.5248의 타깃값 예측
y_predictied = line_fitter.predict(np.array([[1.5248]]))
y_predictied
```

```python
# 회귀 계수 : coef_ 속성
line_fitter.coef_
```

```python
# 절편 확인 : y = 6 + 4X (절편 : 6)
line_fitter.intercept_
```

```python
# 단순 선형 회귀 시각화
plt.plot(X, y, 'o')
plt.plot(X, line_fitter.predict(X))
plt.show()
```

## LinearRegression 클래스 
LinearRegression 클래스  - Ordinary Least Squares  
사이킷런의 linear_model 모듈은 매우 다양한 종류의 선형 기반 회귀를 클래스로 구현해서 제공함  
LinearRegression 클래스는 예측값과 실제값의 RSS를 최소화해서 구현한 클래스이다.  
fit()메서드로 X, y 배열을 입력받아서 predict() 예측하고

## 선형 회귀의 다중공선성 문제  
일반적으로 선형 회귀는 입력 피처의 독립성에 많은 영향을 받음  
피처간의 상관계수가 매우 높은 경우 분산이 매우 커져서 오류에 매우 민감하게 됨  
일반적으로 상관관계가 매우 높은 피처가 많은 경우  
독립적인 중요한 피처만 남기고 제거하거나 규제 적용  
  
다중공선성  
회귀분석에서 독립변수들 간에 강한 상관관계가 나타나는 문제  
  
상관관계가 높은 피처들  
다른 피처에서 파생된 피처

## 회귀 평가 지표

평가 지표 | 설명
:-|:-
**MSE(Mean Squared Error)** | 실제값과 예측값의 차이를 제곱해서 평균한 것  
**RMSE(Root Mean Squard Error)** | MES 값이 오류의 제곱을 구하므로 실제 오류 평균보다 더 커지는 특성이 있으므로 MES에 루트를 씌운 것 <br> MSE에 제곱근을  씌워서 계산하는 함수로 직접 만들어서 사용함 <br> 값이 작을수록 성능이 좋음  
**R제곱** | 분산 기반으로 예측 성능 평가. 실제값의 분산 대비 예측값의 분산 비율을 지표로 하며 1에 가까울수록 예측 정확도가 높다

### 라이브러리 임포트, 데이터 셋 로드

```python
#!pip install seaborn
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
%matplotlib inline

# boston 데이터 셋 로드
boston = load_boston()

# DataFrame으로 변환
bostonDF = pd.DataFrame(boston.data, columns=boston.feature_names)
# bostonDF.head(3)

# boston 데이터셋의 target 주택 가격
# 이를 PRICE 칼럼으로 데이터 프레임에 추가
bostonDF['PRICE'] = boston.target

print('Boston 데이터셋 크기 : ', bostonDF.shape)

bostonDF.head()

# 예측해야 할 값 : PRICE (레이블값, 결정값, 타깃값)
```

```python
bostonDF.info()
# null 값없고 모두 float 형
```

dataset의 feature는 Null 값은 없으며, 모두 float 형임.  
그 다음 각 칼럼(피처값)이 회귀 결과에 미치는 영향이 어느 정도인지 시각화해서 알아보자  
`matplotlib.subplots()`을 이용해 관계를 표현한다

집 값에 크게 영향을 미치는 중요한 요소는

고속도로 접근 용이도 (미국은 차가 있어야 하니까) 생각하면 RAD임    
  
각 칼럼별로 주택 가격에 미치는 영향도를 조사해 보자  
총 8개의 칼럼에 대해 값이 증가할수록 PRICE 갓이 어떻게 변하는지 확인하자  
  
Seaborn의 regplot() 이용하면 각 ax마다 칼럼과 PRICE의 관계를 표현합니다.

```python
# 2개의 행과 4개의 열을 가진 subplots를 이용. axs는 4x2개의 ax를 가짐.
fig, axs = plt.subplots(figsize=(16,8) , ncols=4 , nrows=2)
lm_features = ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']
for i , feature in enumerate(lm_features):
    row = int(i/4)
    col = i%4
    # 시본의 regplot을 이용해 산점도와 선형 회귀 직선을 함께 표현
    sns.regplot(x=feature , y='PRICE',data=bostonDF , ax=axs[row][col])
```

**결과**  
다른 feature보다 RM과 LSTAT의 PRICE 영향도가 가장 두드러지게 나타난다.    
RM(방 개수)는 양 방향의 선형성(Positive Linearity)이 가장 크다.    
즉, 방의 크기가 클수록 가격이 증가하는 모습을 확연히 보여준다.    
LSTAT(하위 계층 비율)는 음 방향으로 선형성(Negative Linearity)이 가장 크다.    
이제 LinearRegression 클래스를 이용하여 보스턴 주택 가격의 회귀 모델을 만들어 학습/예측을 수행한 후 MSE와 R2 Score를 측정해보자  

### LinearRegression 클래스를 이용해서 보스턴 주택 가격의 회귀 모델 생성
1. train_test_split()을 이용해서 학습/테스트 데이터셋 분리
2. fit() 학습, predict() 예측 수행
3. metrics 모듈의 mean_squared_error() 이용해서 MSE 측정
4. r2_score() 이용해서 R2 Score 측정

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 타깃값 데이터 셋 만들기
y_target = bostonDF['PRICE']
# 피처값 데이터 셋 만들기
# 타깃 제외한 나머지 칼럼들
X_data = bostonDF.drop(['PRICE'], axis=1, inplace=False)

X_train , X_test , y_train , y_test = train_test_split(X_data , y_target ,test_size=0.3, random_state=156)

# LinearRegression으로 학습/예측/평가 수행
lr = LinearRegression()
lr.fit(X_train, y_train)
y_preds = lr.predict(X_test)

mse = mean_squared_error(y_test, y_preds) # 예측 오류값
rmse = np.sqrt(mse) # 사이킷런이 지원하지 않기 때문에 np.sqrt() 이용

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('Variance score : {0:.3f}'.format(r2_score(y_test, y_preds)))
```

MSE : 실제값과 예측값의 차이를 제곱해서 평균한 것   
RMSE : MSE에 루트 씌운 값 (낮을 수록 성능 좋음)  
R제곱(r2_scor): 분산 기반으로 예측 성능 평가  
실제값의 분사 대비 예측값의 분산 비율 1에 가까울수록 예측 정확도가 높음  
  
- MSE : 17.297
- RMSE : 4.159
- Variance score : 0.757
나쁜 수치는 아니지만 만족할만한 수치는 아님
  
이제 생성한 주택가격 모델의 intercept(절편)과 coefficients(회귀 계수)값을 보겠습니다.  
절편은 LinearRegression 객체의 intercept_ 속성에, 회귀 계수는 coef_ 속성에 값이 저장돼 있습니다.  

### 주택 가격 모델의 절편과 회귀 계수 값 보기

```python
# LinearRegression으로 생성한 주택가격 모델의
# 절편과 회귀 계수 값 확인
# 절편 : intercept_ 속성에 저장
# 회귀 계수 : coef_ 속성에 저장

print('절편 값 : ', lr.intercept_)
print('회귀 계수 값 : ', np.round(lr.coef_, 1))
```

**교차 검증 수행**  
5개의 폴드 세트에서 corss_val_score()를 이용해 교차 검증 수행  
  
- MSE 구하는 방법  
corss_val_score()의 인자로 'neg_mean_squared_error' 로 설정하면 음수값 반환  
사이킷런의 지표 평가 기준은 높은 지표값일수록 좋은 모델인데 반해  
회귀는 MSE 값이 낮을수록 좋은 회귀 모델이므로  
사이킷런의 metric 평가 기준에 MSE를 부합시키기 위해  
반환된 음수값에 -1을 곱해서 양의 값으로 변환해서 사용  
  
  
- RMSE 구하는 방법  
사이킷런에서 RMSE는 제공하지 않으므로 MSE 수치 결과를 RMSE로 변환할 필요 있음  
변환된 MSE 값에 넘파이 sqrt() 함수를 적용해서 구함  

### 5개의 폴드 세트에서 교차 검증 수행

```python
# 5개의 폴드 세트에서 교차 검증 수행
# MSE와 RMSE 측정

from sklearn.model_selection import cross_val_score

# 5 Fold 셋으로 MSE 구함
neg_mse_scores = cross_val_score(lr, X_data, y_target, scoring="neg_mean_squared_error", cv=5)

# RMSE = MSE에 -1 곱한 후 sqrt() 적용
rmse_scores = np.sqrt(-1 * neg_mse_scores)
avg_rmse = np.mean(rmse_scores)

print('5 folds의 개별 Negative MSE scores : ', np.round(neg_mse_scores, 2))
print('5 folds의 개별 RMSE scores : ', np.round(rmse_scores, 2))
print('5 folds의 평균 RMSE : {0:.3f}'.format(avg_rmse))
```

**결과**  
좀 더 정확한 평가를 위해서 교차 검증을 수행했는데  
5개 폴드 세트에 대해서 교차 검증을 수행한 결과, 평균 RMSE는 약 5.836이 나옴  
반환된 값을 보면 모두 음수임을 알 수 있음  
RMSE값이 약간 더 높게 나타남 만족할 만한 수치는 아님
