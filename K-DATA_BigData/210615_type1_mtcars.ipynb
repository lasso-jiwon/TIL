## 데이터 불러오기

```python
import pandas as pd

mtcars_df = pd.read_csv("mtcars.csv")
mtcars_df.head()
```

```python
# 프레임의 첫 번째 컬럼이 숫자가 아닌 문자열이므로 제거한 후 mtcars_num_df 에 저장
mtcars_num_df = mtcars_df.iloc[:,1:]
mtcars_num_df.head()
```

## 최소최대 척도 (Min-Max Scale)로 변환

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(mtcars_num_df)
mtcars_scaled = scaler.transform(mtcars_num_df)
print(type(mtcars_scaled), mtcars_scaled[:10])
# 데이터 타입이 array임 df 로 바꿔주기
```

```python
mtcars_scaled_df = pd.DataFrame(mtcars_scaled, columns=mtcars_num_df.columns)
mtcars_scaled_df.head()
```

## qsec가 0.5보다 큰 값을 가지는 데이터의 수 구하기

```python
up_df = mtcars_scaled_df[mtcars_scaled_df.qsec>0.5]
up_df.head()
```

```python
answer = len(up_df)
print(answer)
```
